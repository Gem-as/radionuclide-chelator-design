```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem import Descriptors
from rdkit.Chem import Descriptors3D
from rdkit.Chem import rdMolDescriptors
import seaborn as sns
```

```python
from SmilesPE.pretokenizer import atomwise_tokenizer
from SmilesPE.tokenizer import *
import codecs
```

```python
df = pd.read_csv('/content/drive/MyDrive/Информация о связывании медицинского радионуклида различными молекулами.csv')
```

```python
df = df.sort_values(by='lgK', ascending=False).iloc[:105]
```

```python
s = list(df['smiles'])
```

```python
tokens = []
```

```python
for smile in s:
  t = atomwise_tokenizer(smile)
  for i in t:
    if i not in tokens:
      tokens.append(i)
```

```python
spe_vob = codecs.open('/content/drive/MyDrive/SPE_ChEMBL.txt')
spe = SPE_Tokenizer(spe_vob)
```

```python
for smile in s:
  t = spe.tokenize(smile).split(' ')
  for i in t:
    if i not in tokens:
      tokens.append(i)
```

```python
len(tokens)
```

```python
from itertools import combinations
```

```python
generations = []
comb = list(combinations(tokens, 4))
for c in comb:
  r = ''.join([''.join(i) for i in c])
  m = Chem.MolFromSmiles(r,sanitize=False)
  if m != None:
    generations.append(r)
```

```python
len(generations)
```

```python
gen = pd.DataFrame(generations, columns=['smiles'])
gen.to_csv('1.csv')
```

```python
gens = []
```

```python
for smile in s:
  t = spe.tokenize(smile).split(' ')
  for i in range(2, 15):
    comb = list(combinations(t, i))
    for c in comb:
      r = ''.join([''.join(i) for i in c])
      m = Chem.MolFromSmiles(r,sanitize=True)
      if m != None:
        gens.append(r)
```

```python
len(gens)
```

```python
temp2 = []
for i in gens:
  if i not in s:
    temp2.append(i)
```

```python
len(temp2)
```

```python
ge = pd.DataFrame(temp2, columns=['smiles'])
ge.to_csv('4.csv')
```
