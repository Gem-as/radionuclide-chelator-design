```python
symbols = ["C", "O", "N", "P", "S"]

def number_of_atoms(atom_list, df):
    for i in atom_list:
        df['num_of_{}_atoms'.format(i)] = df['mol'].apply(lambda x: len(x.GetSubstructMatches(Chem.MolFromSmiles(i))))

def first(smile):
  for i in symbols:
    smile = smile.replace(i, "")
  for i in smile:
    if i.isalpha():
      return False
  return True

def second(smile):
  uniq = 0
  cur_s = df[df["smiles"] == smile]
  for i in symbols:
    # print(cur_s['num_of_{}_atoms'.format(i)].iloc[0])
    if cur_s['num_of_{}_atoms'.format(i)].iloc[0] > 0:
      uniq += 1
  if uniq >= 2:
    return True
  return False

def third(smile):
  lim = ["O", "N", 'P', "S"]
  cnt_lim = 0
  cur_s = df[df["smiles"] == smile]
  for i in lim:
    cnt_lim += cur_s['num_of_{}_atoms'.format(i)].iloc[0]
  if cnt_lim <= 12:
    return True
  return False

def check_all(df):
  cnst = len(df)
  ok_smiles = [1 for i in range(cnst)]
  df['mol'] = df['smiles'].apply(lambda x: Chem.MolFromSmiles(x))
  df["smiles"] = df['mol'].apply(lambda x: Chem.MolToSmiles(x, isomericSmiles=True, canonical=True))
  number_of_atoms(symbols, df)

  for i in range(cnst):
    if not first(df["smiles"]):
      ok_smiles[i] = 0

  for i in range(cnst):
    if not second(df["smiles"]):
      ok_smiles[i] = 0

  for i in range(cnst):
    if not third(df["smiles"]):
      ok_smiles[i] = 0

  for i in range(cnst):
    if df["smiles"][i] in list(df_all["smiles"]):
      ok_smiles[i] = 0

  for i in range(cnst):
    if Chem.Descriptors.ExactMolWt(df["mol"][i]) > 500:
      ok_smiles[i] = 0

  for i in range(cnst):
    if sascorer.calculateScore(df["mol"][i]) >= 5:
      ok_smiles[i] = 0

  fps = [fpgen.GetFingerprint(x) for x in df["mol"]]
  al_TanimotoSimilarity = 0
  cnt_iter = 0
  for i in range(cnst):
    for j in range(i + 1, cnst):
      cnt_iter += 1
      al_TanimotoSimilarity += DataStructs.TanimotoSimilarity(fps[i],fps[j])
  aver_tanimoto = al_TanimotoSimilarity / cnt_iter
  return [ok_smiles.count(0), ok_smiles, aver_tanimoto]
```

```python
h = check_all(df)
print(f'''Неподходящих {h[0]} ''')
print(f'''tanimoto: {h[2]}''')
```
